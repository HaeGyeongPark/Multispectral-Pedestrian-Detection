```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_bounding_boxes(bboxes, img_width, img_height):
    fig, ax = plt.subplots(1, figsize=(10, 6))

    # 이미지 크기 설정 (배경은 흰색)
    ax.set_xlim(0, img_width)
    ax.set_ylim(0, img_height)
    ax.set_facecolor('white')

    # 바운딩 박스 그리기
    for i, bbox in enumerate(bboxes):
        x, y, width, height = bbox
        
        # 바운딩 박스 추가
        rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        
        # 바운딩 박스 중심점 추가
        ax.plot(x + width / 2, y + height / 2, 'bo')

        # 바운딩 박스 번호 추가
        ax.text(x, y, f'Box {i+1}', fontsize=12, color='black', bbox=dict(facecolor='white', alpha=0.5))
    
    plt.gca().invert_yaxis()  # 좌표축을 이미지와 동일하게 하기 위해 y축을 뒤집습니다.
    plt.show()

# 예시 바운딩 박스 (x, y, width, height) 좌표
bboxes = [
    (561.0,230.75,23.0,50.5),  # 첫 번째 바운딩 박스
    (552.0, 200.0, 22.0, 54.0),  # 두 번째 바운딩 박스
]

# 이미지 크기 (가로, 세로)
img_width = 640
img_height = 480

# 바운딩 박스 그리기
draw_bounding_boxes(bboxes, img_width, img_height)
```
